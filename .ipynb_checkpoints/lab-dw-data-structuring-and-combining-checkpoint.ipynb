# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
demographics = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv")
demographics.drop(columns=["unnamed:_0"], inplace = True)
pd.set_option('display.max_columns', None)
demographics
```

```python
demographics.columns
demographics.dtypes
```

```python
dem_pivot = pd.pivot_table(demographics, index = 'sales_channel', columns = "coverage", values = "customer_lifetime_value", aggfunc='sum').round(2)
dem_pivot["Total Revenue"] = dem_pivot.sum(axis = 1)
dem_pivot
```

```python
#as you can see, agents produce the largest revenue, and the web service produces the lowest amount.
```

```python
lifetime_pivot = pd.pivot_table(demographics, index = "education", columns = "gender", values = "customer_lifetime_value", aggfunc = "mean").round(2)
lifetime_pivot
```
